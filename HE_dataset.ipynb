```python
import os
import cv2

def histogram_equalization(input_dir, output_dir):
    # Create the output directory if it doesn't exist
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    # Iterate over all files in the input directory
    for filename in os.listdir(input_dir):
        # Check if the file is an image
        if filename.lower().endswith(('.png', '.jpg', '.jpeg')):
            # Read the image
            img_path = os.path.join(input_dir, filename)
            img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)

            # Apply histogram equalization
            equalized_img = cv2.equalizeHist(img)

            # Save the equalized image to the output directory
            output_path = os.path.join(output_dir, filename)
            cv2.imwrite(output_path, equalized_img)

            print(f"Processed image: {filename}")

# Provide the input and output directory paths
input_directory = 'archive/colored_images/Severe'
output_directory = 'archive/HE_dataset/Severe'

# Call the histogram_equalization function
histogram_equalization(input_directory, output_directory)
```
