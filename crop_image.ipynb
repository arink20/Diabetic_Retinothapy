```python
import os
import random 
import matplotlib.pyplot as plt
import cv2
```

```python
image_dir = "Datasets/Messidor/all_images"

tif_files = [file for file in os.listdir(image_dir) if file.endswith(".tif")]
```

```python
random_tif = random.choice(tif_files)
image_path = os.path.join(image_dir, random_tif)

image = cv2.imread(image_path)
resized_image = cv2.resize(image,(512,512))
```

```python
crop_height  = 345
crop_width = 345

# Calculate cropping indices
y1 = (resized_image.shape[0] - crop_height) // 2
y2 = y1 + crop_height
x1 = (resized_image.shape[1] - crop_width) // 2
x2 = x1 + crop_width

# Perform the center crop
center_cropped = resized_image[y1:y2, x1:x2]
```

```python
plt.figure(figsize=(10, 8))

plt.subplot(2, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Original Image")

plt.subplot(2, 2, 2)
plt.imshow(cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB))
plt.title("Resized Image (224x224)")

plt.subplot(2, 2, 3)
plt.imshow(cv2.cvtColor(center_cropped, cv2.COLOR_BGR2RGB))
plt.title("Center Cropped Image")

plt.tight_layout()
plt.show()
```
